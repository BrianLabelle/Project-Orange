```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import linregress
```

```python
build_permits = pd.read_csv("../CensusBuildingPermitsCum2000_2018.csv")

build_permits.head()
```

```python
years = build_permits.iloc[:,0]
values = build_permits.iloc[:,4]
units = build_permits.iloc[:,1]

plt.plot(years,values,color="green")

plt.xlim(2000, 2018)

plt.title('Metro Houston Residential Building Permits 2000-2018')
plt.xlabel("Years")
plt.ylabel("Cumulative Value of Permits in Billions")
```

```python
plt.plot(years,units,color="blue")

plt.xlim(2000, 2018)

plt.title('Metro Houston Residential Building Permits 2000-2018')
plt.xlabel("Years")
plt.ylabel("Number of Permits")
```

```python
fig, ax = plt.subplots()

(slope, intercept, _, _, _) = linregress(years, values)
fit = slope * years + intercept
ax.plot(years, fit,'r--')
ax.bar(years,values)

plt.xlim(2000, 2018)

plt.title('Metro Houston Residential Building Permits 2000-2018')
plt.xlabel("Years")
plt.ylabel("Value of Permits in Billions")
```

```python
avecompaction = pd.read_csv("../Toalave_compbyyear.csv")

avecompaction
```

```python
subsidence = avecompaction.iloc[:,1]

# fig, ax = plt.subplots()

plt.figure(figsize=(10, 10))

plt.subplot(2, 1, 1)
plt.plot(years, values, 'r')
plt.title('Metro Houston Cumulative Building Permits 2000-2018')
plt.xlabel("Years")
plt.ylabel("Value of Permits in Billions of Dollars")
plt.xlim(2000, 2018)


plt.subplot(2, 1, 2)
plt.plot(years, subsidence, 'g')
plt.title('Metro Houston Average Subsidence/Compaction 2000-2018')
plt.ylabel('Average Subsidence (feet)')
plt.xlabel("Years")
plt.xlim(2000, 2018)

plt.tight_layout()
plt.show()

plt.savefig("../permitsvscompaction.png")
```
