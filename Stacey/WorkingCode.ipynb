```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import linregress
```

```python
build_permits = pd.read_csv("../CensusBuildingPermits2000_2018.csv")

build_permits.head()
```

```python
years = build_permits.iloc[:,0]
values = build_permits.iloc[:,3]
units = build_permits.iloc[:,1]

plt.plot(years,values,color="green")

plt.xlim(2000, 2018)

plt.title('Metro Houston Residential Building Permits 2000-2018')
plt.xlabel("Years")
plt.ylabel("Value of Permits in Billions")
```

```python
plt.plot(years,units,color="blue")

plt.xlim(2000, 2018)

plt.title('Metro Houston Residential Building Permits 2000-2018')
plt.xlabel("Years")
plt.ylabel("Number of Permits")
```

```python
fig, ax = plt.subplots()

(slope, intercept, _, _, _) = linregress(years, values)
fit = slope * years + intercept
ax.plot(years, fit,'r--')
ax.bar(years,values)

plt.xlim(2000, 2018)

plt.title('Metro Houston Residential Building Permits 2000-2018')
plt.xlabel("Years")
plt.ylabel("Value of Permits in Billions")
```
